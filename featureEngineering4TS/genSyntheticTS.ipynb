```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# generate simple synthetic timeseries dataset: alpha * x + ampl * sin(freq * x) + noise_level
np.random.seed(123)

total_points = 365 * 3
freq = 30 * np.pi / total_points
ampl = 1
noise_level = 1
alpha = 0.001


# generate time serie: alpha * x + ampl * sin(freq * x) + noise_level
time = np.arange(total_points)
data = alpha * time + ampl * np.sin(freq * time) + np.random.normal(0, noise_level, total_points)
```

```python
df_ts = pd.DataFrame(data, columns=['Value'])
df_ts['DateTime'] = pd.date_range(start='2020-01-01', periods=total_points, freq='D')
df_ts = df_ts[['DateTime', 'Value']]  # Reordering columns for readability
df_ts.head()
```

```python
#Plot the synthetic time series
plt.figure(figsize=(10, 6))
plt.plot(df_ts['DateTime'], df_ts['Value'], label='Synthetic Time Series')
plt.xlabel('DateTime')
plt.ylabel('Value')
plt.title('Synthetic Time Series Data')
plt.legend()
plt.show()
```
